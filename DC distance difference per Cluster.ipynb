```python
import pandas as pd
import geopandas as gpd
import folium
from shapely.geometry import Point

# Step 1: Load your data
file_path = 'test.csv'
clusters = pd.read_csv(file_path)

# Check the column names
print(clusters.columns)

# Step 2: Calculate Cluster Centroids
centroids = clusters.groupby(['Territory', 'cluster_code'])[['latitude', 'longitude']].mean().reset_index()

# Step 3: Assign Colors to Territories
territory_colors = {'Territory1': 'blue', 'Territory2': 'green', 'Territory3': 'red'}  # Add more as needed

# Step 4: Create a GeoDataFrame
geometry = [Point(xy) for xy in zip(centroids['longitude'], centroids['latitude'])]
gdf = gpd.GeoDataFrame(centroids, geometry=geometry)

# Step 5: Create Cluster Polygons
gdf['buffer'] = gdf.buffer(0.1)  # Buffer around each point to create polygons

# Step 6: Create Folium Map
m = folium.Map(location=[gdf['latitude'].mean(), gdf['longitude'].mean()], zoom_start=10)

for index, row in gdf.iterrows():
    folium.GeoJson(
        data=row['buffer'].__geo_interface__,
        tooltip=f"Territory: {row['Territory']}<br>Cluster: {row['cluster_code']}",
        style_function=lambda x, color=territory_colors.get(row['Territory'], 'gray'): {'fillColor': color}
    ).add_to(m)

# Save the map as an HTML file
m.save('cluster_map.html')
```

```python
import pandas as pd
import geopandas as gpd
import folium
from shapely.geometry import Point, Polygon

# Step 1: Load your data
file_path = 'test.csv'
clusters = pd.read_csv(file_path)

# Step 2: Calculate Cluster Centroids
centroids = clusters.groupby(['Territory', 'cluster_code'])[['latitude', 'longitude']].mean().reset_index()

# Step 3: Assign Colors to Territories
territory_colors = {'Syokimau': 'red', 'Eastlands': 'blue', 'Embakasi': 'green', 'Kahawa': 'purple', 'Kiambu': 'pink', 'Thika': 'orange', 'Waiyaki Way': 'yellow', 'Rongai': 'black', 'Machakos': 'brown', 'Makadara': 'white'}  # Add more as needed

# Step 4: Create a GeoDataFrame for Clusters
geometry_clusters = [Point(xy) for xy in zip(centroids['longitude'], centroids['latitude'])]
gdf_clusters = gpd.GeoDataFrame(centroids, geometry=geometry_clusters)

# Step 5: Create a GeoDataFrame for Territories
territory_polygons = []
for territory, color in territory_colors.items():
    territory_clusters = gdf_clusters[gdf_clusters['Territory'] == territory]
    if not territory_clusters.empty:
        territory_polygon = Polygon(territory_clusters['geometry']).buffer(0.05)  # Buffer around clusters
        territory_polygons.append({'geometry': territory_polygon, 'Territory': territory, 'color': color})

gdf_territories = gpd.GeoDataFrame(territory_polygons)

# Step 6: Create Folium Map
m = folium.Map(location=[gdf_clusters['latitude'].mean(), gdf_clusters['longitude'].mean()], zoom_start=10)

# Step 7: Plot Territories on the Map
for index, row in gdf_territories.iterrows():
    folium.GeoJson(
        data=row['geometry'].__geo_interface__,
        tooltip=f"Territory: {row['Territory']}",
        style_function=lambda x, color=row['color']: {'fillColor': color, 'color': color, 'weight': 1}
    ).add_to(m)

# Step 8: Plot Clusters on the Map
for index, row in gdf_clusters.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"Territory: {row['Territory']}<br>Cluster: {row['cluster_code']}",
        icon=folium.Icon(color=territory_colors.get(row['Territory'], 'gray'))
    ).add_to(m)

# Save the map as an HTML file
m.save('cluster_map.html')
```
